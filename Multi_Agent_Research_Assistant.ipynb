```python
from transformers import pipeline

research_agent = pipeline(
    "text-generation",
    model="gpt2"
)

summary_agent = pipeline(
    "text-generation",
    model="gpt2"
)

analysis_agent = pipeline(
    "text-generation",
    model="gpt2"
)

print("Multi-Agent Research Assistant Ready")
```

```python
topic = input("Enter Research Topic: ")

# Research Agent
research_prompt = f"Give research information about {topic}"

research_output = research_agent(
    research_prompt,
    max_length=150,
    num_return_sequences=1
)

research_text = research_output[0]['generated_text']

print("\nResearch Agent Output:\n")
print(research_text)

# Summary Agent
summary_prompt = f"Summarize this:\n{research_text}"

summary_output = summary_agent(
    summary_prompt,
    max_length=100,
    num_return_sequences=1
)

summary_text = summary_output[0]['generated_text']

print("\nSummary Agent Output:\n")
print(summary_text)

# Analysis Agent
analysis_prompt = f"Analyze this research:\n{summary_text}"

analysis_output = analysis_agent(
    analysis_prompt,
    max_length=120,
    num_return_sequences=1
)

analysis_text = analysis_output[0]['generated_text']

print("\nAnalysis Agent Output:\n")
print(analysis_text)
```
